# 02 — Exploratory Data Analysis

Verifies the dataset before training:
- Class balance
- Sample image grids
- Pixel value distribution
- Corrupt image detection

**Run after 01_data_preparation.ipynb.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

train_df = pd.read_csv('/kaggle/working/train.csv')
print(train_df['label'].value_counts())
```

```python
# --- Class balance chart ---
train_df['label'].value_counts().plot(kind='bar', color=['#ef4444','#22c55e'])
plt.xticks([0, 1], ['Non-nature (0)', 'Nature element (1)'], rotation=0)
plt.title('Class balance')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# --- Sample 5x5 positive images ---
pos_samples = train_df[train_df.label == 1].sample(25, random_state=42)['path'].tolist()
fig, axes = plt.subplots(5, 5, figsize=(12, 12))
for ax, path in zip(axes.flat, pos_samples):
    try:
        img = Image.open(path).convert('RGB').resize((112, 112))
        ax.imshow(img)
    except Exception:
        ax.text(0.5, 0.5, 'Error', ha='center')
    ax.axis('off')
plt.suptitle('Positive samples (nature elements)')
plt.tight_layout()
plt.show()
```

```python
# --- Sample 5x5 negative images ---
neg_samples = train_df[train_df.label == 0].sample(25, random_state=42)['path'].tolist()
fig, axes = plt.subplots(5, 5, figsize=(12, 12))
for ax, path in zip(axes.flat, neg_samples):
    try:
        img = Image.open(path).convert('RGB').resize((112, 112))
        ax.imshow(img)
    except Exception:
        ax.text(0.5, 0.5, 'Error', ha='center')
    ax.axis('off')
plt.suptitle('Negative samples (non-nature)')
plt.tight_layout()
plt.show()
```

```python
# --- Detect corrupt images ---
all_paths = pd.read_csv('/kaggle/working/train.csv')['path'].tolist()
corrupt = []
for p in all_paths[:500]:  # spot check first 500
    try:
        Image.open(p).verify()
    except Exception:
        corrupt.append(p)

print(f'Corrupt images in sample: {len(corrupt)}')
if corrupt:
    print(corrupt[:5])
```
